#### Question 2: Find the most voted question today with at least a score of 5 and tagged with 'python'. 

```python
# your code here
SITE = StackAPI('stackoverflow')
questions2 = SITE.fetch('questions', fromdate='2020-09-01', todate='2020-09-30',min=5, tagged='python', sort='votes')

data2 = pd.DataFrame(questions2)

data2 = flatten(data2, ['items'])
data2 = flatten(data2, ['items_owner'])
data2
```

```python

question2_score = pd.DataFrame(data2 ['items_score'])

question2_score
```
